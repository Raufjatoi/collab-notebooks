```python
## Datatypes in csv
data=('col1,col2,col3\n'
      'x,y,1,3\n'
      'a,b,c,4\n'
       'e,f,e')
df=pd.read_csv(StringIO(data))
df.info()
```

```python
df=pd.read_csv(StringIO(data),dtype='object')
```

```python
df=pd.read_csv(StringIO(data),dtype='string')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
data = ('a,b,c,d\n'
        '1,2,3,4\n'
        '5,6,7,8\n'
        '9,10,11')
df=pd.read_csv(StringIO(data),dtype={'a':int,'b':float,'c':int})
df.info()
```

```python
df.dtypes
```

```python
data = ('index,a,b,c\n'
        '4,apple,bat,5.7\n'
        '8,orange,cow,10')
```

```python
pd.read_csv(StringIO(data),index_col=0)
```

```python
data = ('index,a,b,c\n'
        '4,apple,bat,5.7\n'
        '8,orange,cow,10')
pd.read_csv(StringIO(data),usecols=['a','b','c'],index_col=0)
```
